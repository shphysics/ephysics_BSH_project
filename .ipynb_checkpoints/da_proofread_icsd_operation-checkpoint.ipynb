```python
def extract_components(string):
    # Regular expression pattern to match letters and numbers
    pattern = r'([A-Za-z]+|\.\d+(?:\.\d+)?|\d+(?:\.\d+)*)'

    components = re.findall(pattern, string)

    return components
```

```python
comp_num_lst_lsts = []
for stuff in merged_full_in_icsd_no_spaces:
    comp_num_lst_lsts.append(extract_components(stuff))
```

```python
def filter_strings(strings, comp_num_lst_lsts):
    # Regular expression pattern to match
    pattern = r'^(?!.*0\.0$).*$'

    filtered_strings = []

    for idx, string in enumerate(strings):
        stuff = comp_num_lst_lsts[idx]
        if len(stuff) <= 4 and (stuff[-1] != '0.0' or stuff[-1] != '0.' or stuff[-1] != '0'):
            print(idx,string)
        elif re.match(pattern, string) or len(stuff) >= 6:
            filtered_strings.append(string)
        else:
            print(idx, string)


    return filtered_strings
```

```python
proof_icsd_removing_single_comp = filter_strings(merged_full_in_icsd_no_spaces, comp_num_lst_lsts)
```

```python
def remove_decimal_repeated_strings(strings, comp_lst):
    # Regular expression pattern to match the specified patterns
    pattern = r'.*?(\d+\.\d+)(?:\1)+.*?'
    filtered_strings = []
    things_catch_to_remove = []
    for i, string in enumerate(strings):
        stuff = comp_lst[i]
        for comp in stuff:
            if comp.count('.') >= 2:
                print(i, string, comp)
                things_catch_to_remove.append(string)
                continue
        if re.match(pattern, string):
            print(i, string)
            continue
        else:
            filtered_strings.append(string)
    
    
    return [elem for elem in filtered_strings if elem not in things_catch_to_remove]
```

```python
proof_icsd_removing_repeated_decimal = remove_decimal_repeated_strings(proof_icsd_removing_single_comp, comp_num_lst_lsts_removing_single_comp)
```

```python
len(proof_icsd_removing_repeated_decimal)
```

```python
count_period_remove = 0
def remove_dot_at_end(strings, count_period_remove):
    updated_strings = []
    for string in strings:
#         print(string[-1])
        if string.endswith('.'):
            updated_strings.append(string[:-1])
            count_period_remove += 1
        else:
            updated_strings.append(string)
    return updated_strings
```

```python
proof_icsd = remove_dot_at_end(proof_icsd_removing_repeated_decimal, count_period_remove)
print(count_period_remove)
```
